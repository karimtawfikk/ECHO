```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen-0.6B-Embeddings")
model = AutoModel.from_pretrained("Qwen/Qwen-0.6B-Embeddings")

inputs = tokenizer(["your text here"], return_tensors="pt", padding=True, truncation=True)
with torch.no_grad():
    embeddings = model(**inputs).last_hidden_state.mean(dim=1)
```
